## Objective

The objective of this project is to:
- Read a CSV file into a Pandas DataFrame
- Inspect the dataset
- Compute summary statistics
- Filter rows
- Save filtered results

Dataset Used: Titanic Dataset (train.csv)

```python
import pandas as pd
```

## 1. Reading the CSV File

In this step, we load the Titanic dataset into a Pandas DataFrame.

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("train.csv")
df.head()
```

## 2. Inspecting the Dataset

We examine the last rows, data types, and structure of the dataset.

```python
df.tail()
```

```python
df.dtypes
```

```python
df.info()
```

## 3. Summary Statistics

We compute statistical measures such as mean, median, minimum, maximum, and count.

```python
df.describe()
```

```python
print("Mean Age:", df["Age"].mean())
print("Median Fare:", df["Fare"].median())
print("Minimum Age:", df["Age"].min())
print("Maximum Fare:", df["Fare"].max())
print("Total Passengers:", df["PassengerId"].count())
```

## 4. Filtering the Dataset

Here, we filter passengers who survived (Survived = 1).

```python
survived = df[df["Survived"] == 1]
survived.head()
```

## 5. Saving Filtered Data

The filtered dataset is saved into a new CSV file.

```python
survived.to_csv("survived_passengers.csv", index=False)
```

## 6. Data Visualization

We create a bar chart to visualize the survival distribution.

```python
import matplotlib.pyplot as plt

survival_counts = df["Survived"].value_counts()

survival_counts.plot(kind="bar")

plt.title("Survival Count of Passengers")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")

plt.show()
```
